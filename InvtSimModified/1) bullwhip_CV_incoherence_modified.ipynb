### CV and Bullwhip Effect Ratio

```python
import numpy as np
import pandas as pd
from tqdm import tqdm
```

```python
path = 'E:/yjz/Extension for hts/JayCode/Models/'

orders = pd.read_pickle(f"all_order.pkl")#.drop(columns = 'forecasts')
uid    = pd.read_pickle(f"{path}721future_28.pkl")[['unique_id','ds']]
prd_idx= pd.read_pickle(f"prd_idx.pkl")
# ao = pd.read_pickle("actual_order.pkl")
```

```python
%%time
lst = orders['name'].unique().tolist()
df  = orders.copy()
frdr_ = pd.concat([pd.concat([uid,  df[df['name'] == i].reset_index(drop = True)], axis = 1)for i in tqdm(lst)])#.reset_index()# ao,

dff = frdr_.copy()
split_cols = dff['unique_id'].str.split('/', expand=False)

dff['product'] = pd.DataFrame([ i[-1] for  i in tqdm(split_cols)]) 
dff['level'] = pd.DataFrame([len(i) - 3 for i in tqdm(split_cols)])
```

```python
dff
```

```python
var_ = dff.groupby(['unique_id', 'name', 'product', 'level']).var()
lst2 = var_.columns.tolist()
var_[lst2[-13:]] = var_[lst2[-13:]].div(var_['true_demand'], axis=0)
# var_['forecasts'] = var_['forecasts'].div(var_['true_demand'], axis = 0)
```

```python
var_
```

```python
### .xs('FOODS_1_001', level='product')
df2 = var_[np.isfinite(var_).all(axis=1)] # dropna and inf value
uids = df2.index.get_level_values('unique_id').unique()#.tolist()
#uids.str.split('/').str[-1].tolist()
vlist = []
for i in tqdm(uids.str.split('/').str[-1].tolist()):
    df = df2.xs(f'{i}', level='product').groupby(['level', 'name']).mean()#['level', 'product', 'name']
    vlist.append(df) 
```

```python
pd.concat(vlist).groupby(['level', 'name']).median().iloc[:,3:].to_csv("bwe.csv")
```

```python
pd.concat(vlist).groupby(['level', 'name']).median().iloc[:,3:] # volatility
```

## Mean bullwhip summary

```python
bullwhip_mean_summary = (
    pd.concat(vlist)
    .groupby(['level', 'name'])
    .mean()
    .iloc[:, 3:]
    .reset_index()
)
bullwhip_mean_summary
```
